```python
##Class Distribution (Fraud vs Non-Fraud)
fraud_count = df['Class'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=fraud_count.index, y=fraud_count.values,hue=fraud_count.index)
plt.title("Fraud vs Non-Fraud Transaction Count")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Number of Transactions")
plt.show()
#value_counts()
##fraud_count.index → Class names (0 and 1)
##fraud_count.values → Their counts
##Tells us how many frauds and non-frauds exist..index
##Used to label the x-axis with 0 and 1..values
##Used to get the number of transactions for each class.
```

```python
##5.Transaction Amount Distribution
plt.figure(figsize=(7,5))
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
df.sort_values(by='Amount').value_counts(df[df['Amount']<=5000]['Amount'])
```

```python
##6. Compare Fraud vs Non-Fraud Behaviour (EDA)
##6.1 Amount Comparison
plt.figure(figsize=(7,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Fraud vs Non-Fraud Transaction Amount")
plt.show()
##Interpretation:Fraudulent transactions often have distinct patterns (either very small or very large).


```

```python
##6.2 Time-Based Behaviour
##Dataset has a Time column in seconds.
plt.figure(figsize=(10,5))
sns.kdeplot(df[df.Class==0]['Time'], label="Non-Fraud")
sns.kdeplot(df[df.Class==1]['Time'], label="Fraud")
plt.legend()
plt.title("Fraud vs Non-Fraud Transaction Time Pattern")
plt.show()
##df[df['Class']==1]['Amount']**
##Select the *Amount* column for fraud transactions.
##df[df['Class']==1]['Amount']**
##Select the *Amount* column for fraud transactions.
```

```python
print("non-fraud","\n",df[df['Class']==0]['Time'],"\n","fraud","\n",df[df['Class']==1]['Time'])
```

```python
##Correlation Study (Find Important Fraud Indicators)
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
##Money Flow Visualization (Using Plotly)
##This shows how money moves across amounts.
fig = px.histogram(df, x="Amount", color="Class", nbins=50, title="Money Flow: Fraud vs Non-Fraud")
fig.show()
```

```python
## Identify Red Flags Using EDA Only
##9.1 Unusual High Amount Transactions
high_amount = df[df['Amount'] > df['Amount'].quantile(0.99)]
high_amount[high_amount['Class'] == 1].head()
```

```python
##9.2 Transactions in Rare Time Windows
rare_times = df[(df["Time"] < 500) | (df["Time"] > 85000)]
rare_times[rare_times["Class"] == 1].head()
```

```python
##9.3 Top features linked to fraud
fraud_corr = df[df['Class'] == 1].corr()['Class'].sort_values(ascending=False)
fraud_corr.head(10)
```

```python
##PCA Visualization (2D Plot)
##Helpful to show clusters of fraud.
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Scale data
X = df.drop('Class', axis=1)
X = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
pca_comp = pca.fit_transform(X)
df_pca = pd.DataFrame({"PC1": pca_comp[:,0], "PC2": pca_comp[:,1], "Class": df["Class"]})
fig = px.scatter(df_pca, x='PC1', y='PC2', color='Class', title="PCA Visualization of Fraud Patterns")
fig.show()
```
